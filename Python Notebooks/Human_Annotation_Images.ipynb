```python
import glob
import pandas as pd
from PIL import Image
import psutil
```

```python
df = pd.read_csv('/Users/Meghna/Desktop/twitter_race_tweets_noRT.csv')
df.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
df.drop(["a"], axis=1, inplace=True)
df
```

```python
df_ss = df[df['race'] == 1]
df_ss
# df_ss = df[df['race'] == 0]
# df_ss
```

```python
user_ids= df_ss['user_id'].to_list()
print(len(user_ids))
user_ids = set(user_ids)
```

```python
print(len(user_ids))
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
human_annotate = pd.DataFrame(columns = ('user_id','decision'))
path = '/Users/Meghna/Desktop/twitter_test_images/'
files = glob.glob(path)
count = 0
for ids in user_ids:
    imgplot = plt.imshow(mpimg.imread(path + str(ids) + '.jpg'))
    plt.ion()
    plt.show()
    #img = Image.open(path + str(ids) + '.jpg')
    #img.show()
    print('Enter Decision')
    decide = input('Yes/ No/ Not sure :')
    print('You entered ', decide)
    new_row = {'user_id': ids, 'decision': decide}
    human_annotate = human_annotate.append(new_row, ignore_index=True)
    count += 1
    print(count)
    plt.close()
```

```python
human_annotate.to_csv('/Users/Meghna/Desktop/human_annotations.csv')
```
